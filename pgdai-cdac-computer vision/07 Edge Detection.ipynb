```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
img = cv2.imread('draw.png')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGRA2RGB))
```

#### Canny edge detection

```python
# Perform edge detection (low threshold, high threshold)
edges = cv2.Canny(img, 50, 150)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Edge Detected')
plt.imshow(edges, cmap = 'gray')
```

#### Laplacian Edge Detection

```python
img = cv2.imread('draw.png')
```

```python
# Perform Laplacian edge detection
kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

# src, ddepth (-1 ignore), kernel
edges = cv2.filter2D(img, -1, kernel)
```

```python
kernel.shape
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Edge Detected')
plt.imshow(edges, cmap = 'gray')
```

```python
cv2.imwrite('mask.png', edges)
```
