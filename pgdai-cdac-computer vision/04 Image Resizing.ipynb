```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
img = cv2.imread('dog.jpg')
```

```python
plt.imshow(img)
```

```python
newImg = cv2.resize(img, (0,0), fx=0.75, fy=0.75)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(newImg)
```

```python
newImg = cv2.resize(img, (550, 350))
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(newImg)
```
